```python
from scipy import stats
import pandas as pd

# Our FPdataset.csv is composed of dak.csv, heart.csv, data.csv, Heart_Disease_Prediction.csv
# heart.csv
dfheart = pd.read_csv('heart.csv')
# data.csv
dfdata = pd.read_csv('data.csv', na_values = "?").fillna("0")
# the new columns list 
old_names = list(dfdata.columns)
new_names = list(dfheart.columns)
dfdata.columns = new_names
# we use the dfdata dataframe as the orginal one to append the dfheart on 
df = dfdata.append(dfheart, ignore_index=True)
```

```python
# Heart_Disease_Prediction.csv
dfheart_pred = pd.read_csv(' Heart_Disease_Prediction.csv')
dfheart_pred.columns = new_names
# append the dfheart_pred on existing one
df = df.append(dfheart_pred, ignore_index=True)
df.replace("Absence", "0",inplace = True)
df.replace("Presence", "1",inplace = True)
```

```python
# dak.csv
dfbak = pd.read_csv('bak.csv', na_values = "?").fillna("0")
dfbak.columns = new_names
# append the dfbak on the existing one 
df = df.append(dfbak, ignore_index=True)
# save the dataset as FPdataset.csv
df.drop_duplicates()
df.to_csv("FPdataset.csv")
```
